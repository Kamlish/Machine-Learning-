```python
x = np.array([1,2,3,4,5,6])
```

```python
x
```

```python
y = 2*x + 1
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-lable")
plt.ylabel("y-label")
```

```python
plt.plot(x,y,color="r", linestyle=":", linewidth=2)
plt.title("Line Plot")
plt.xlabel("x-lable")
plt.ylabel("y-label")
```
