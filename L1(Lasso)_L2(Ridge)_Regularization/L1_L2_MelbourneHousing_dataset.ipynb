```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# read dataset
df = pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
# Exploratory Analysis
df.nunique()
```

```python
df.shape
```

```python
col_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
             'Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize',
             'BuildingArea','Price']
df = df[col_to_use]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
col_to_fill_zero = ['Distance','Bedroom2','Bathroom','Propertycount','Car']
df[col_to_fill_zero] = df[col_to_fill_zero].fillna(0)
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Hot EnConding 
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
X = df.drop('Price', axis='columns')
y = df[['Price']]
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
# Model
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
# To Avoid Overfitting 
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=55, max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=55, max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
