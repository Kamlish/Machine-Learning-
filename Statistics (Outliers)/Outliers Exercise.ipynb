```python
df = pd.read_csv("bp.csv")
```

```python
df.price.describe()
```

```python
min_threshold, max_threshold= df.price.quantile([0.001,0.999])
```

```python
outliers = df[(df.price<min_threshold) | (df.price>max_threshold)]
outliers.shape
```

```python
no_outliers = df[(df.price>min_threshold) & (df.price<max_threshold)]
```

```python
no_outliers.shape
```

```python
no_outliers
```

```python
mean = no_outliers.price.mean()
```

```python
std = no_outliers.price.std()
```

```python
left_d = mean - 4*std
```

```python
right_d = mean + 4*std
```

```python
right_d
```

```python
std_outliers= no_outliers[(no_outliers.price>right_d) | (no_outliers.price<left_d) ]
```

```python
std_outliers
```

```python
std_no_outliers = no_outliers[(no_outliers.price>left_d) & (no_outliers.price<right_d) ]
```

```python
std_no_outliers
```
